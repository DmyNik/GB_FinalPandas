```python
# 1.1. Скачать следующие данные: kc-house-data и laptop_price

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df_house = pd.read_csv('task_1-3/kc-house-data.csv')
df_laptop = pd.read_csv('task_1-3/laptop_price.csv')

# df_house.head()
# df_laptop.head()

df_house.info()
# df_laptop.info()
```

```python
# 1.2. Изучите стоимости недвижимости

data = df_house.groupby(['price']).size().reset_index(name='count')

# print(data.head(15))

plt.figure(figsize=(8, 4.5))

plt.scatter(data['price'], data['count'])

# plt.boxplot([data['price']])

plt.title('Cтоимости недвижимости')
plt.xlabel('Стоимость недвижимости, млн $')
plt.ylabel('Количество домов, шт')

plt.rc('grid', linestyle='--', color='gray', linewidth=1)
plt.grid (True)

delta_y = 20
y_ticks = []
i = 0
y_ticks.append(i)

while i < data['count'].max():
  i += delta_y
  y_ticks.append(i)

delta_x = 500000
x_ticks = []
i = 0
x_ticks.append(i)

while i < data['price'].max():
  i += delta_x
  x_ticks.append(i)

plt.yticks(y_ticks);
plt.xticks(x_ticks);

# price_str = str('Cтоимости недвижимости: от ' + str(df_house['price'].min()) + '$ до '
#             + str(df_house['price'].max()) + '$, '
#             + 'средняя стоимость - ' + str(round(df_house['price'].mean(), 2)) + str() + '$')

# print(price_str)

#
# Выводы:
#
# На этом графике мы видим, что начиная от самой минимальной стоимости количество домов резко 
# растет и в районе отметки 500 000 $ находится самое большое число домов. Однако после данных
# цен по мере увеличения стоимости их количество начинает резко снижаться.
# В районе 1,0 млн $ - 1,5 млн $ тренд на снижение количества начиет замедляться
# и уменьшение их числа после 1,5 млн $ становится плавным.
# Однако в районе 6,0 млн $ - 6,5 млн $ мы совсем не видим домов,
# а в районе 7,0 млн $ - 8,0 млн $ мы наблюдаем совсем небольшое их число и это
# самые дорогие дома в этой таблице.
```

```python
# 1.3. Изучите распределение квадратуры жилой

data = df_house.groupby(['sqft_living']).size().reset_index(name='count')

# print(data.head(15))

plt.figure(figsize=(9, 4.5))

plt.scatter(data['sqft_living'], data['count'])

plt.title('Квадратура жилой площади')
plt.xlabel('Жилая площадь, кв. м')
plt.ylabel('Количество домов, шт')

plt.rc('grid', linestyle='--', color='gray', linewidth=1)
plt.grid (True)

delta_y = 20
y_ticks = []
i = 0
y_ticks.append(i)

while i < data['count'].max():
  i += delta_y
  y_ticks.append(i)

delta_x = 1000
x_ticks = []
i = 0
x_ticks.append(i)

while i < data['sqft_living'].max():
  i += delta_x
  x_ticks.append(i)

plt.yticks(y_ticks);
plt.xticks(x_ticks, rotation = 50);

liv_sqr_str = str('Распределение квадратуры жилой площади: от ' + str(df_house['sqft_living'].min()) + ' кв.м до '
            + str(df_house['sqft_living'].max()) + ' кв.м, '
            + 'средняя жилая площадь - ' + str(round(df_house['sqft_living'].mean(), 2)) + str() + ' кв.м')

print(liv_sqr_str)

#
# Выводы:
#
# На графике мы видим, что значительная часть домов имеет жилую площадь примерно от 500 до 4000 кв. м.
# Не очень большая, но все-таки довольно заметная часть домов имеет площадь от 4000 до 8000 кв. м и
# совсем немного домов имеют жилую площадь от 8000 до 14000 кв. м.
# Также мы можем заметить, что очень многие дома в диапазоне от 4000 до 8000 кв. м
# имеют срандартизированные показатели жилой площади и, вероятно эти дома построены по типовым проектам.
# Больше всего таких домов находится в районе отметки 1500 кв. м. Но, вместе с тем во всех 
# категориях домов по жилой площади встречаются дома, у которых показатель жилой площади является 
# практически уникальным.
```

```python
# 2.2. Изучите распределение этажей домов

atribusts = ['price', 'yr_built', 'view', 'sqft_living']

data = df_house.groupby(['floors']).size().reset_index(name='count')

# print(data.head(15))

sns.jointplot(x=data['floors'], y=data['count'], kind='reg')
plt.title('Количество домов по этажам', y=1.25)
plt.grid (True)

for str in atribusts:
    sns.jointplot(x=df_house['floors'], y=df_house[str], kind='reg')
    plt.title(str, y=1.25)
    plt.grid (True)

#
# Выводы:
#
# На графиках мы видим, что количество домов в данной таблице уменьшается
# с увеличением этажности дома. Также, обращает на себя внимание то, что
# домов с нецелой этажностью меньше, чем домов с целой.
# Цена дома увиличивается с числом этажей.
# Среди высохих домов больше домов новой постройки.
# Красивый вид из окна практически не завит от количества этажей в доме.
# Жилая площадь растет по мере увеличения этажности дома.
```

```python
# Условие 3: 3 задача
# Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
# Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

df_house_corr = df_house.select_dtypes(include=['float64', 'float32', 'float16', 'int64', 'int32', 'int16'])
df_house_corr.drop('id', axis=1, inplace=True)
corr_matrix = df_house_corr.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.4] = 0
corr_matrix

#
# Выводы:
#
# Исходя из представленной матрицы корреляций делаем выводы о том, что на стоимость недвижимости значимое влияние
# оказывают такие следующие параметры дома:
# - Количество ванных комнат;
# - Жилая площадь;
# - Вид из окна;
# - Качество конструкции и дизайна дома;
# - Площадь дома, которая находится выше земли;
# - Жилая площадь у 15 соседей.
```

```python
# Сделаем представление данных более наглядным

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, linewidths=0.5, cmap='coolwarm')
plt.tick_params(axis='both', which='major', labelsize=10, labelbottom = False, bottom=False, top=False, labeltop=True)
plt.xticks(rotation = 90)
plt.title('Матрица корреляций')
```

```python
# Покажем другие графики кореляций цены дома с наиболее существенными параметрами: Ящики с усами

atribusts = ['bathrooms', 'view', 'grade']

for str in atribusts:
    plt.figure(figsize=(9, 5))
    plt.grid(True)
    sns.boxplot(x='price', y=str, data=df_house, showfliers=False, orient='h')
    plt.title(str, y=1.075)
    
#
# Выводы:
#
# Данные графики хорошо визуализируют выбранные параметры,
# однако, первый график из-за разбивки на большое число категорий 
# становится трудночитаемым.
```

```python
# Покажем другие графики кореляций цены дома с наиболее существенными параметрами: Столбчатые диаграммы

atribusts = ['bathrooms', 'view', 'grade']

for str in atribusts:
    plt.figure(figsize=(8, 4))
    plt.grid(True)
    size = df_house[str].max()
    if (size % 2 != 0): size += 1
    # print(size)
    sns.histplot(df_house[str], bins=int(size))
    plt.title(str, y=1.075)
    
#
# Выводы:
#
# В связи с тем, что количество домов по разным категориям параметров на данных графиках
# слишком неравномерно, то данные графики в целом тоже малоинформативны.
# Однако, они могут оказаться полезны, если нужно визуально сравнить и проанализировать
# отдельные категории параматров.
```

```python
# Покажем другие графики кореляций цены дома с наиболее существенными параметрами: Многорядовые столбчатые диаграммы

dataset = pd.DataFrame()

max_value = df_house['price'].max()

grade = 1
temp_value = max_value

while (temp_value > 10): 
    temp_value /= 10
    grade *= 10

max_value =  int(max_value/grade) if ((max_value/grade) % 1 == 0) else int(max_value/grade) + 1  

data = pd.Series()

arr1 = [] # Цены
arr2 = [] # Ванны
arr3 = [] # Вид
arr4 = [] # Категории

for i in range(0, max_value):
   arr1.append((i + 1) * grade)
   arr2.append(df_house[(df_house['price'] >= i * grade) & (df_house['price'] <= (i+1) * grade)]['bathrooms'].mean())
   arr3.append(df_house[(df_house['price'] >= i * grade) & (df_house['price'] <= (i+1) * grade)]['view'].mean())
   arr4.append(df_house[(df_house['price'] >= i * grade) & (df_house['price'] <= (i+1) * grade)]['grade'].mean())

dataset['price'] = arr1
dataset['bathrooms'] = arr2
dataset['view'] = arr3
dataset['grade'] = arr4

plt.figure(figsize=(8, 4))

n_ticks = np.arange(len(dataset['price'])) # Определяем кол-во делений (тиков)
                                           # делений столько - сколько объектов в таблице

offset = -0.333 # Определяем сдвиг столбиков
w = 0.27 # Определяем ширину столбца

# Добавляем сдвиг к количеству делений
plt.bar(n_ticks + offset*0.5, dataset['bathrooms'], width=w)
plt.bar(n_ticks + offset*1.5, dataset['view'], width=w)
plt.bar(n_ticks + offset*2.5, dataset['grade'], width=w)

plt.title('Зависимости количества ванн, вида из окна и конструкции дома от его цены')
plt.xlabel('Род занятий')
plt.ylabel('Среднее кол-во в ценовой категории')
plt.legend(['Количество ванн', 'Категория вида из окна', 'Категория конструкции'])
plt.xticks(n_ticks, dataset.price, rotation = 30); # Добавляем метки делений

#
# Выводы:
#
# Данный график получился довольно компакным и наглядным, однако на нем не сразу можно понять,
# что все три столбика имеют отношение к домам одной стоимости, а не то, что зеленый - 
# к более дешевым домам, а синий к более дорогим.
```

```python
# Покажем другие графики кореляций цены дома с наиболее существенными параметрами: Круговые диаграммы

atribusts = ['bathrooms', 'view', 'grade']

for str in atribusts:
    data = df_house[str].value_counts()
    data.index = df_house[str].unique()
    plt.figure(figsize=(10, 5))
    plt.pie(data, autopct='%1.lf%%')
    plt.title('Распределение домов по: ' + str)
    plt.legend(data.index)
    
#
# Выводы:
#
# Первый графи не слишком удобоно воспринимать, видимо нужно что-то сделать с маленькими значениями.
# Остальный два графика вполне информативны.
```
